
# Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading Data


```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
print(data)
```

```python
#We are converting the column names so that they have no spaces.
```

```python
X = pd.DataFrame(data=data.data,columns=[i.replace(' ','_') for i in data.feature_names])
X['target'] = data.target
```

# Training the Model


```python
from sklearn.model_selection import train_test_split
```

```python
y = X['target']
X = X.drop('target',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Training using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegressionModel = LogisticRegression()
```

```python
logisticRegressionModel.fit(X_train,y_train)
```

```python
predictions = logisticRegressionModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
#With logistic regression, we are getting pretty great results.Now let us try with SVM.
```

Training using SVM

```python
from sklearn.svm import SVC
```

```python
supportVectorModel = SVC()
```

```python
supportVectorModel.fit(X_train,y_train)
```

```python
svm_predictions = supportVectorModel.predict(X_test)
```

```python
print(classification_report(y_test,svm_predictions))
```

```python
print(confusion_matrix(y_test,svm_predictions))
```

```python
#We are getting such terrible results because the features are not normalised.
```

First Improvement - Normalisation

```python
#Normalising training data

X_train_min = X_train.min()
range_train = X_train.max() - X_train.min()
X_train_scaled = (X_train - X_train_min) / range_train
```

```python
#Normalising testing data

X_test_min = X_test.min()
range_test = X_test.max() - X_test.min()
X_test_scaled = (X_test - X_test_min) / range_test
```

```python
logisticRegressionModel.fit(X_train_scaled,y_train)
```

```python
predictions = logisticRegressionModel.predict(X_test_scaled)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
supportVectorModel.fit(X_train_scaled,y_train)
```

```python
svm_predictions = supportVectorModel.predict(X_test_scaled)
```

```python
print(classification_report(y_test,svm_predictions))
```

```python
print(confusion_matrix(y_test,svm_predictions))
```

```python
#As a result of normalisation, the results from the logistic regression model were not affected much. 
#But the result from the SVM model improved a lot.
```

Second Improvement - Choosing the best C and gamma

```python
param_grid = {'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridSearchModel = GridSearchCV(SVC(),param_grid,verbose = 3,refit = True)
```

```python
gridSearchModel.fit(X_train_scaled,y_train)
```

```python
gridSearchModel.best_params_
```

```python
grid_predictions = gridSearchModel.predict(X_test_scaled)
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
#We got furthur improvement using the optimum C and gamma for SVM method
```
